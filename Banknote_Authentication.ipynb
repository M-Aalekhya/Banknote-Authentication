```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

data = pd.read_csv('Banknote_authentication_dataset.csv')
```

```python
V1 = data['V1']
V2 = data['V2']

V1_Mean = np.mean(V1)
V1_Std = np.std(V1)
print(V1_Mean, V1_Std)

V2_Mean = np.mean(V2)
V2_Std = np.std(V2)
print(V2_Mean, V2_Std)
```

```python
plt.xlabel('V1 Values')
plt.ylabel('V2 Values')
plt.scatter(V1, V2, s = 40, alpha = 0.3)
plt.savefig('fig1')
plt.show()
```

```python
km_res = KMeans(n_clusters = 2).fit(data)
clusters = km_res.cluster_centers_
plt.scatter(V1, V2, s = 40, alpha = 0.3)
plt.scatter(clusters[:,0], clusters[:,1], s= 100)
plt.savefig('fig2')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2).fit(data.iloc[:,:-1]) 
data['cluster'] = kmeans.labels_
```

```python
fig, ax = plt.subplots(1,2,figsize=(10,5))
ax[0].scatter(data['V1'], data['V2'], c=data['cluster'], cmap='rainbow') 
ax[0].set_title('K-means Clustering Results')

ax[1].scatter(data['V1'], data['V2'], c=data['cluster'], cmap='rainbow')
ax[1].set_title('Actual Classes')
plt.savefig('fig3')
plt.show()
```

```python
figs, axs = plt.subplots(2,5, figsize=(15,6))

for i in range(10):
    kmeans = KMeans(n_clusters=2).fit(data.iloc[:,:-1])
    data['cluster'] = kmeans.labels_
    
    axs[i%2,i//2].scatter(data['V1'], data['V2'], c=data['cluster'], cmap='rainbow')
    axs[i%2,i//2].set_title('Run '+str(i+1))
    plt.savefig('fig' + str(i+4))
```
